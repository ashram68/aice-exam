## 문제 7. 잔당 IQR 이상치 처리 (오류 정정)  `8점`

잔당(`residual_sugar`)에서 이상치가 발견되었습니다. IQR 기준으로 처리하세요.
- **(7-1)** 아래 코드에 오류가 있습니다. 코드의 오류를 정정하세요.
  - 대상: `df` / IQR = Q3 − Q1, 상한 = Q3 + 1.5·IQR, 하한 = Q1 − 1.5·IQR
  - 잔당이 상한 초과 또는 하한 미만인 행을 삭제 → `df_temp`에 저장
- **(7-2)** 이상치 처리로 삭제된 행은 총 몇 개인가요?


```python
# (7-1) 아래 코드의 오류를 정정하고 실행하세요
q1 = df["residual_sugar"].quantile(0.25)
q3 = df["residual_sugar"].quantile(0.75)
iqr = q1 - q3                                 # ← 오류: IQR = Q3 - Q1
upper = q3 + 1.5 * iqr
lower = q1 - 1.5 * iqr
df_temp = df.drop( df["residual_sugar"] > upper )   # ← 오류: 이상치 index만 drop, 상·하한 모두 반영


# (7-2) 여기에 답안을 입력하세요(실행 불필요)
# 답안 = 
```
