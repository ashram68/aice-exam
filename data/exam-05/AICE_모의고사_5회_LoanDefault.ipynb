## 문제 6. 결측치 처리 (오류 정정)  `8점`

결측치는 데이터 유형에 따라 다른 대표값으로 대체합니다. 수치형은 중앙값, 범주형은 최빈값으로 처리하세요.
- **(6-1)** 아래 코드에 에러가 나고 있습니다. 코드의 오류를 정정하세요.
